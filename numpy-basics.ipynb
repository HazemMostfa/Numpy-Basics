**Accessing/Changing specific elements, rows, columns, etc**


```python
a=np.array([[1,3,5,7,9],[2,4,6,8,10]])
print(a)
```

```python
#get spesific elemant 
a[1,3]
```

```python
# get spesific row
a[0,:]
```

```python
# get spesific column
a[:,4]
```

```python
# getting little more fancy
a[1,0:-1:4]
```

```python
# change elemant
a[:,3]=[15,16]
```

**Initializing Different Types of Arrays**

```python
# All ones matrix
np.ones((5,3))
```

```python
# Any other number
np.full((3,3),50)
```

```python
# Any other number (full like)
np.full_like(c,20)
```

```python
# Random decimal numbers
np.random.rand(3,2)
```

```python
np.identity(3)
```

```python
# repeat an array
arr=np.array([[1,2,3]])
ar1=np.repeat(arr,3,axis=0)
ar1
```

**Reorganizing Arrays**

```python
b=np.array([[1,3,5,7],[2,4,6,8]])
b
a=b.reshape((4,2))
a
```

```python
# Vertically stacking vectors
v1=np.array([1,3,5,7])
v2=np.array([2,4,6,8])
np.vstack([v1,v2,v1])
```

```python
# Horizontal stack
h1=np.zeros((3,3))
h2=np.ones((3,3))
np.hstack([h1,h2])
```
